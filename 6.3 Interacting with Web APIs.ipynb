# Interacting with Web APIs

```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
res = requests.get(url) # You need to connected with Internet.
```

```python
res
```

```python
data = res.json()
```

```python
data[0]['title']
```

```python
issues = pd.DataFrame(data, columns=['number', 'title', 'labels', 'state'])
```
